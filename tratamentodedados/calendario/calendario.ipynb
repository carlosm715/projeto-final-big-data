```python
def dia_semana(date):
    datanum = date.weekday()
    
    if datanum == 6:
        return "Domingo"
    elif datanum == 0:
        return "Segunda-feira"
    elif datanum == 1:
        return "Terça-feira"
    elif datanum == 2:
        return "Quarta-feira"
    elif datanum == 3:
        return "Quinta-feira"
    elif datanum == 4:
        return "Sexta-feira"
    else:
        return "Sábado"
   
```
